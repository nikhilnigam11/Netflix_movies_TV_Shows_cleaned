# **Missing Values**
1.   director column has 30 % missing values
2.   cast column has 9.3 % missing values
3.   country column has 9.4 % missing values

The missing values in the director column has 2634 NaN values which correspond with almost 30 percents of total data in that column. So, with just dropping the **NaN** values, we will lose lots of data, instead filling the NaN values with empty string is a good choice.

```python
df.dropna(subset=['date_added','rating','duration'], inplace = True)
```

```python
print('\033[1m<-----------Missing Values Percentage----------->\033[1m')
pd.DataFrame({'Total missing values':df.isna().sum(),
              'Percentage':round((df.isna().sum()/len(df))*100,2)})
```

```python
#replacing the NaN in director with the 'name not provided'

df.fillna({'director':'Director Name Not Provided'}, inplace=True)
df.head()
```

```python
#replacing the NaN in cast with the 'Name not provided'

df.fillna({'cast':'Name Not Provided'}, inplace=True)
df.head()
```

```python
print('\033[1m<-----------Missing Values Percentage----------->\033[1m')
pd.DataFrame({'Total missing values':df.isna().sum(),
              'Percentage':round((df.isna().sum()/len(df))*100,2)})
```

# **Cleaning the Comma Separated Values in the director and cast column**

```python
# Saperating the comma saperated values in the director column
df['director'] = df['director'].str.split(',')
df = df.explode('director')
df['director'] = df['director'].str.strip()
df.head()
```

```python
# Saperating the comma saperated values in the cast column
df['cast'] = df['cast'].str.split(',')
df = df.explode('cast')
df['cast'] = df['cast'].str.strip()
df.head()
```

```python
dfn=df[df.duplicated()]
dfn.shape
```

10 duplicate rows are found. Now Dropping them to get rid of duplicate values

```python
df.drop_duplicates(inplace=True)
```

```python
dfn=df[df.duplicated()]
dfn.shape
```

Duplicate Rows dropped

```python
# Saperating the comma saperated values in the 'listed_in' column
df['listed_in'] = df['listed_in'].str.split(',')
df = df.explode('listed_in')
df['listed_in'] = df['listed_in'].str.strip()
df.head()
```

```python
# Saperating the comma saperated values in the 'listed_in' column
df['country'] = df['country'].str.split(',')
df = df.explode('country')
df['country'] = df['country'].str.strip()
df.head()
```

```python
df['release_year'] = pd.to_datetime(df['release_year'], format='%Y')
df['release_year'] = df['release_year'].dt.year
df.head()
```

# **leading spaces are observed in the date_added column data so to_datetime is giving error.**

```python
df['date_added'] = df['date_added'].str.strip()
df.head()
```

```python
df['date_added']=pd.to_datetime(df['date_added'])
df.head()
```

```python
df.fillna({'country':'country Name Not Provided'}, inplace=True)
df.head()
```
